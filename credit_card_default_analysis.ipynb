```python
for col in categorical_columns:
    print(f"\n --- Default Rate by {col.upper()} ---\n")
    result = pd.crosstab(df[col], df['default_flag'], normalize = 'index') * 100
    print(result.round(2))
```

```python
# Chi-Square Test of Independence

from scipy.stats import chi2_contingency

cat_columns_not_default = ['sex', 'education', 'marriage']

for col in cat_columns_not_default:
    table = pd.crosstab(df[col], df['default_flag'])
    chi2, p, dof, expected = chi2_contingency(table)
    
    print(f"\n--- Chi-Square Test: {col.upper()} ---")
    print(f"ChiÂ² Statistic = {chi2:.2f}")
    print(f"Degrees of Freedom = {dof}")
    print(f"P-value = {p:.4f}")
    
    if p < 0.05:
        print("Significant relationship with default_flag (reject null hypothesis)")
    else:
        print("Not significant (fail to reject null hypothesis)")
```

```python
print(f"True p-value = {p}")
```

```python
import numpy as np
from scipy.stats import chi2_contingency

def cramers_v(confusion_matrix):
    chi2 = chi2_contingency(confusion_matrix)[0]
    n = confusion_matrix.sum().sum()
    phi2 = chi2 / n
    r, k = confusion_matrix.shape
    return np.sqrt(phi2 / min(k - 1, r - 1))

for col in cat_columns_not_default:
    table = pd.crosstab(df[col], df['default_flag'])
    v = cramers_v(table)
    print(f"CramÃ©râ€™s V for {col}: {v:.4f}")
```

### **Categorical Analysis Observation:**

- Even though all three variables are statistically significant (Chi-Square p < 0.05), their real-world influence on default behavior is very small.

- These categorical variables do impact default likelihood, but only slightly.

- Most of the predictive power likely lies in the numerical variables, such as: repayment history, credit limit, billing amount, and payments made.

```python
from scipy.stats import ttest_ind

numerical_columns = [
    'limit_bal', 'age',
    'pay_0', 'pay_2', 'pay_3', 'pay_4', 'pay_5', 'pay_6',
    'bill_amt1', 'bill_amt2', 'bill_amt3', 'bill_amt4', 'bill_amt5', 'bill_amt6',
    'pay_amt1', 'pay_amt2', 'pay_amt3', 'pay_amt4', 'pay_amt5', 'pay_amt6'
]

print(f"{'Feature':<15} {'P-Value':<12} {'Significant (<0.05)':<20}")
print('-'*50)

for col in numerical_columns:
    group1 = df[df['default payment next month'] == 1][col]  # Defaulters
    group2 = df[df['default payment next month'] == 0][col]  # Non-defaulters
    
    t_stat, p_val = ttest_ind(group1, group2, equal_var=False)
    
    is_significant = 'Yes' if p_val < 0.05 else 'No'
    print(f"{col:<15} {p_val:<12.4e} {is_significant:<20}")
```

```python
grouped_means = df.groupby('default payment next month')[numerical_columns].mean().T
grouped_means.columns = ['No Default', 'Default']
grouped_means['Difference'] = grouped_means['Default'] - grouped_means['No Default']
grouped_means.round(2)
```

### Statistical Insights: Numerical Variables vs Default

- Customers who defaulted had significantly lower credit limits, lower monthly payments, and more frequent payment delays across all repayment status columns (pay_0 to pay_6).
- Repayment behavior (pay_0 to pay_6) is the Strongest predictor, showing that defaulters consistently fell behind on payments.
- Defaulters also paid ~2000â€“3000 less per month compared to non-defaulters.
- The last 3 months of bill_amt showed no significant difference suggesting that billing history alone isnâ€™t a strong standalone predictor.
